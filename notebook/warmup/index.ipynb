```python
# warmup 1: perceptron

def perceptron_classify(w, b, x):
    return 1*(x@w - b > 0)
    #np.dpt (w, x) == w@x
perceptron_classify([1,2,3], 35, [4,5,6])
```

```python
# warmup 2: convexity

from matplotlib import pyplot as plt 
import numpy as np

# line 1 points
x1 = np.linspace(-.5,1.5,100)
y1 = -np.log(x1)
# plotting the line 1 points 
plt.plot(x1, y1, label = "f(z) = -ln(z)")
  
# line 2 points
x2 = np.linspace(-.5,1.5,100)
y2 = -np.log(1-x2)
# plotting the line 2 points
plt.plot(x2, y2, label = "g(z) = -ln(1-z)")

# giving a title to my graph
plt.title('Practice Plotting')
  
# show a legend on the plot
plt.legend(loc='upper right')
  
# function to show the plot
plt.show()
```
